```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Names of the .csv files
file1 = 'neocando-Land Bank, vacant Residential.csv'
file2 = 'neocando-Popul_Pov_Unemp_Crime.csv'
file3 = 'neocnado-Fore, Sales, med prices, parcels by neighborhood.csv'
```

```python
# Read each .csv into a dataframe
df1 = pd.read_csv(file1, encoding="ISO-8859-1")
df2 = pd.read_csv(file2, encoding="ISO-8859-1")
df3 = pd.read_csv(file3, encoding="ISO-8859-1")
```

```python
# Preview df1
df1.head()
```

```python
# Preview df2
df2.head()
```

```python
# Preview df3
df3.head()
```

```python
# Merge df1 and df2 using an inner join
merge_table_temp = pd.merge(df1, df2, on='Neighborhood')
merge_table_temp
```

```python
# Merge merge_table_temp and df3 using an inner join.  Ready for analysis!
indicators = pd.merge(merge_table_temp, df3, on='Neighborhood')
indicators.head()
```

```python
# Neighborhoods with high Land Bank involvement
high_lb = indicators.iloc[[1,7,15,24,31],:]
high_lb.head()
```

```python
# Neighborhoods with low Land Bank involvement
low_lb = indicators.iloc[[2,16,19,26,30],:]
low_lb.head()
```

```python
# concatenate the high and low 
frames = [high_lb, low_lb]
indicators_high_low = pd.concat(frames)
indicators_high_low.reset_index
indicators_high_low.head(10)
```

```python
indicators_high_low.to_csv('indicators.csv') 
```
